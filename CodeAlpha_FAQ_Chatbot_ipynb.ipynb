```python
df = pd.read_csv("faqs.csv")

df.head()
```

```python
def preprocess(text):

    text = text.lower()

    text = text.translate(
        str.maketrans('', '', string.punctuation)
    )

    return text
```

```python
questions = df["Question"].apply(preprocess)

vectorizer = TfidfVectorizer()

tfidf_matrix = vectorizer.fit_transform(questions)
```

```python
def chatbot(user_question):

    user_question = preprocess(user_question)

    user_vector = vectorizer.transform([user_question])

    similarity_scores = cosine_similarity(
        user_vector,
        tfidf_matrix
    )

    best_match_index = similarity_scores.argmax()

    confidence = similarity_scores[0][best_match_index]

    if confidence < 0.20:
        return (
            "Sorry, I couldn't understand your question.",
            confidence
        )

    answer = df.iloc[best_match_index]["Answer"]

    return answer, confidence
```

```python
print("=" * 50)
print("        AI FAQ CHATBOT")
print("=" * 50)
print("Type 'exit' to stop.\n")

while True:

    user_input = input("You: ")

    if user_input.lower() == "exit":
        print("Bot: Goodbye!")
        break

    answer, confidence = chatbot(user_input)

    print(f"Bot: {answer}")
    print(f"Confidence Score: {confidence:.2f}\n")
```
